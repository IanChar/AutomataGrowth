# Finding Probability Model for Substring Matching Suffix (Started April 16 2017)

## Set up

Let $A_i$ be the event that there is some string in the generalized language $G$ of length $\ell = |G|$ such that the string has a substring starting at index $i$ of length $k$ that matches with one of the prefixes in $G$. The overall goal is then to find,

$$P\left(\cup_{i=1}^{\ell - k + 1} A_i\right)$$

## Probability of Non-empty Intersection

Let $M_{i, j}$ be the event that $G[i] \cap G[j] \neq \emptyset$. In other words, $M_{i, j}$ is the event that there is at least one pair of strings that share the same letter in index $i$ and $j$. Note that where $a = |\Sigma|$...

$$P(M_{i, j}) = 1 - \frac{3^a - 2^{a + 1} + 1}{\left(2^a - 1\right)^2}$$

This probability assumes that $G$ was picked such that each possible non-empty subset of the alphabet has equal chance of showing up for an index. 

Why this probability? To justify this considier the number of ways there are of having $G[i] \cap G[j] = \emptyset$. For each letter in $\sigma \in \Sigma$, $\sigma$ could either be in the first set and not the second, the second and not the first, or neither sets, hence $3^a$. However, by doing this we count the configurations where one of the sets is the null set. Thus we must subtract all configurations where one of the sets is the null set, hence $3^a - 2(2^a)$. Lastly, when this term was subtracted the case where each of the sets are the empty set was double counted, hence $3^a - 2^{a+1} + 1$.

To get the final probability divide this quantity by the total number of possible configurations, $(2^a - 1)^2$, and take the complement.

This expression can be simplified to the following:

$$P(M_{i, j}) = \frac{2^{2a} - 3^a}{\left(2^a - 1\right)^2}$$

Running simulations for validation...

```python
# Constants
TRIALS = 10000
ALPH_SIZE_RANGE = range(2, 50)

def theoretical_prob(a):
    return (2 ** (2 * a) - 3 ** a) / ((2 ** a - 1) ** 2)

def plot_differences():
    # Get theoretical values as listed above.
    theoretical = [theoretical_prob(a) for a in ALPH_SIZE_RANGE]
    # Get simulated values
    simulated = []
    for a in ALPH_SIZE_RANGE:
        alph = [chr(ord('A') + i) for i in range(a)]
        num_matches = 0
        for _ in xrange(TRIALS):
            pairOfSets = [set(aho_test.build_random_string(1, alph)[0]) for _ in range(2)]
            if len(pairOfSets[0].intersection(pairOfSets[1])) > 0:
                num_matches += 1
        simulated.append(num_matches / TRIALS)
    # Plot the two values together and plot the difference.
    plt.figure(1, figsize=(14, 6))
    plt.subplot(121)
    plt.plot(ALPH_SIZE_RANGE, theoretical, label='Theoretical Probabilities')
    plt.plot(ALPH_SIZE_RANGE, simulated, label='Simulated Probabilities')
    plt.legend(loc='upper left')
    plt.title('Probabilities')
    plt.xlabel('Alphabet Size')
    plt.ylabel('Probability')
    
    differences = [simulated[i] - theoretical[i] for i in range(len(theoretical))]
    plt.subplot(122)
    plt.plot(ALPH_SIZE_RANGE, differences)
    plt.title('Simulated minus Theoretical')
    plt.xlabel('Alphabet Size')
    plt.ylabel('Probability')
    plt.show()
    
plot_differences()
```

The theoretical and simulated values seem to agree to a large extent. However, it is weird how there is much higher error for smaller alphabet sizes. This is probably because whith smaller alphabet sizes there is higher variance so it is more likely the simulations will not agree. Furthermore, the error seems to be random and not systematic, further supporting this assertion.

It is also strange that there is an initial dip in the probability before it shoots up and asymptotes to 1.

## Probability of Match

Recall that $A_i$ is the event that there exists some string in $G$ that has a k-length match with a prefix starting at index $i$. Note then that,
$$
A_i = \cap_{j = 0}^{k - 1} M_{i + j, 1 + j}
$$

Under the condition that $i > k$, each of these events will be independent. Why? If $i \leq k$ then there is overlap in the sets that we are matching. Thus the outcome of one event can inform the outcome of another. That being said, if $\ell$ is much greater than $k$ this will almost always be the case. This is becuase out of all possible substrings, the proportion that do not overlap is $\frac{\ell -k + 1}{\ell}$.

Assuming that the match between the substring and prefix do not overlap...
$$
P(A_i) = \left(\frac{2^{2a} - 3^a}{\left(2^a - 1\right)^2}\right)^k
$$

## Possible use of Bonferroni Inequality

Can we use Bonferroni inequality to help approximate the probability of there being a k-length match with a prefix?

Let $\beta_r$ be the set of all possible sets containing $r$ of the events $A_2, \ldots, A_{\ell - k + 1}$. Define $S_r$ to be the following:

$$
S_r = \sum_{B \in \beta_r} P\left(\cap_{A_i \in B} A_i \right)
$$

The Bonferroni inequality then says that...

$$
\sum_{j = 1}^{\textrm{Some even } n} (-1)^{j - 1} S_j \leq P(\cup A_i) \leq \sum_{j = 1}^{\textrm{Some odd } n} (-1)^{j - 1} S_j
$$

Calculating $S_r$ poses a problem. To make this simpler for now, I will assume that the $A_i$s are independent from each other. Note that this is almost certainly not the case. Not only could the substrings overlap, but matching with a prefix tells us some information about the possible prefixes. That being said, if this assumption is taken...

$$
S_r = {{\ell - k + 1}\choose{r}} \left(\frac{2^{2a} - 3^a}{\left(2^a - 1\right)^2}\right)^{rk}
$$

Using this, let's see what happens when we simulate the lower and upper bounds...

```python
# Constants
STRING_LENGTH = 50
MATCH_SIZE = 20
SUM_MAX = 6
ALPHABET_SIZE = 4

# Taken from http://stackoverflow.com/questions/4941753/is-there-a-math-ncr-function-in-python
def ncr(n, r):
    r = min(r, n-r)
    if r == 0: 
        return 1
    numer = reduce(op.mul, xrange(n, n-r, -1))
    denom = reduce(op.mul, xrange(1, r+1))
    return numer / denom

def get_sr(r):
    to_return = ncr(STRING_LENGTH - MATCH_SIZE + 1, r)
    to_return *= ((2 ** (2 * ALPHABET_SIZE) - 3 ** ALPHABET_SIZE) / ((2 ** ALPHABET_SIZE - 1) ** 2)) ** (r * MATCH_SIZE)
    return to_return

def get_lower_and_upper_bounds(bound_max):
    lower = []
    upper = []
    total = 0
    for i in range(1, bound_max + 1):
        if i % 2 == 0:
            total -= get_sr(i)
            lower.append(total)
        else:
            total += get_sr(i)
            upper.append(total)
    return (lower, upper)

def plot_lower_and_upper():
    lower, upper = get_lower_and_upper_bounds(SUM_MAX)
    low_max, high_min = max(lower), min(upper)
    print 'Lower Bound: %f' % low_max
    print 'Upper Bound: %f' % high_min
    
    plt.plot(range(1, len(lower) + 1), lower, label='Lower Bound')
    plt.plot(range(1, len(upper) + 1), upper, label='Upper Bound')
    plt.legend(loc='upper right')
    plt.title('Lower and Upper Bounds for Probability')
    plt.xlabel('Bound Number')
    plt.ylabel('Probability')
    plt.show()
    
plot_lower_and_upper()
```

The upper and lower bound here seem to squish down to the desired probability very quickly (when lower bound is $r = 4$ and when upper bound is $r = 3$). Several question arise: 

1) How accurate is this? We are assuming that fact that $A_i$ are independent and we are assuming that all $M_{i, j}$ are independent even though there are some that overlap with the prefix. 

2) Are there ways to use Bonferroni without making these assumptions?

3) How fast can we expect the upper and lower bound to "squish" down to the probability? 

4) Is there any advantage to doing this over using the complement and doing intersections? Do they give the same probability?

### Question 1: How accurate is this?

```python
# Constants
STRING_LENGTH = 50
MATCH_SIZE = 20
SUM_MAX = 6
ALPHABET_SIZE = 4
TRIALS = 10000

def sim_prob_of_kmatch(trials):
    alph = [chr(ord('A') + i) for i in range(ALPHABET_SIZE)]
    total_matches = 0
    for _ in xrange(trials):
        rand_string = aho_test.build_random_string(STRING_LENGTH, alph)
        # Check if there is a k-size match
        hasMatch = True
        for start_index in range(1, len(rand_string) - MATCH_SIZE):
            for index in range(MATCH_SIZE):
                if len(set(rand_string[start_index + index]).intersection(set(rand_string[index]))) == 0:
                    hasMatch = False
                    break
            if hasMatch:
                total_matches += 1
                break
            else:
                hasMatch = True
    return total_matches / trials

print sim_prob_of_kmatch(TRIALS)
```
